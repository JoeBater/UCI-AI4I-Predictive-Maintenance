```python
import pandas as pd
import numpy as np

from sklearn import tree
import matplotlib.pyplot as plt
```

```python
dirpath = r'C:\Users\joeba\OneDrive\Documents\ML\Data\aps failure uci'

df = pd.read_csv(dirpath+'/aps_failure_training_set.csv', skiprows=20)

df.head(10)
```

```python
df['class'].value_counts()
```

```python
df = df.replace('na', np.NaN)
df = df.fillna(0)

df.isna().sum()
```

```python
d = {'pos': 1, 'neg': 0}
df['class'] = df['class'].map(d)

df = df.astype(float)
```

```python
features = list(df.columns)[1:]
```

```python

X = df[features]
y = df['class']

#dt = tree.DecisionTreeClassifier(min_weight_fraction_leaf=0.01)
#dt = tree.DecisionTreeClassifier(class_weight='balanced')
dt = tree.DecisionTreeClassifier(class_weight={1:2})
dt = dt.fit(X,y)
```

```python
tst = pd.read_csv(dirpath+'/aps_failure_test_set.csv', skiprows=20)

tst = tst.replace('na', np.NaN)
tst = tst.fillna(0)

tst['class'] = tst['class'].map(d)
tst = tst.astype(float)

tst
```

```python
tX = tst[features]
ty = tst['class']
```

```python
pred = dt.predict(tX)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(ty, pred)
cm
```

```python
from sklearn.metrics import classification_report

target_names = ['neg', 'pos']
print(classification_report(ty, pred, target_names=target_names))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=250, random_state=1223, class_weight={1:0.49})
rf = rf.fit(X,y)

rf_pred = rf.predict(tX)

rf_cm = confusion_matrix(ty, rf_pred)
print(rf_cm)
print()
print(classification_report(ty, rf_pred, target_names=target_names))
```
